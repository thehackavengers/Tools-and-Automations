```python
from gdeltdoc import Filters, near, repeat



def fetch_article_link(from_, to_, keyword, domain):
    """ function to fetch article links from gdelt 2.0 api related to India in English langauage"""

    # creating filter
    f = Filters(
        start_date = from_,
        end_date = to_,
        keyword = keyword,
        country = "India",
        language= "English",
        domain = domain,
        tone = "<-1"
    )

    gd = GdeltDoc()

    # Search for articles matching the filters
    df = gd.article_search(f)

    return df 

```

```python
# Fetching news urls for all keywords and news sources

from_ = "2025-04-01"
to_ = "2025-04-30"
df_list = []
domain_list = list(set(news_domains))
for domain in domain_list:
    for keyword in keywords:
        print(f"fetching data for keyword: {keyword} domain: {domain} ")
        df = fetch_article_link(from_, to_, keyword, domain)
        print(f'Fetched {df.shape[0]} records!!')
        df_list.append(df)
        
print("Fetching Done !!")

link_df = pd.concat(df_list)
link_df.to_csv('./apr_2025_links.csv')
print(link_df.shape)
link_df.head()
```
